```python
from rdkit import Chem

PeriodicTable = Chem.GetPeriodicTable()
```

```python
element_symbols = []

for i in range(1,21):
  element_symbols.append(PeriodicTable.GetElementSymbol(i))
```

```python
numero_atomico = [PeriodicTable.GetAtomicNumber(x) for x in element_symbols]
masa_atomica = [PeriodicTable.GetAtomicWeight(x) for x in element_symbols]
nombre_del_elemento = [PeriodicTable.GetElementName(x) for x in range(1,21)]
valencias = [PeriodicTable.GetValenceList(x) for x in range(1,21)]
```

```python
import pandas as pd

data = {'nombre': nombre_del_elemento, 'simbolo': element_symbols,
        'masa_atomica': masa_atomica, 'numero_atomico': numero_atomico,
        'valencias': valencias}
df = pd.DataFrame(data)
df
```

```python
def obtener_electrones_de_valencia(numero_atomico):
    if numero_atomico == 1:
        return 1
    elif numero_atomico == 2:
        return 2
    elif 3 <= numero_atomico <= 10:
        return numero_atomico - 2
    elif 11 <= numero_atomico <= 18:
        return numero_atomico - 10
    elif 19 <= numero_atomico <= 36:
        return numero_atomico - 18
    elif 37 <= numero_atomico <= 54:
        return numero_atomico - 36
    elif 55 <= numero_atomico <= 86:
        return numero_atomico - 54
    elif 87 <= numero_atomico <= 118:
        return numero_atomico - 86
    else:
        return None

try:
    numero_atomico = int(input("Ingresa el número atómico: "))
    electrones_valencia = obtener_electrones_de_valencia(numero_atomico)

    if electrones_valencia is not None:
        print(f"El número de electrones de valencia para el elemento con número atómico {numero_atomico} es: {electrones_valencia}")
    else:
        print("Número atómico fuera de rango.")
except ValueError:
    print("Por favor, ingresa un número entero válido.")
```
